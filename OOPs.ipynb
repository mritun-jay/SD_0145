```python
class Student:
  # Attributes (Data)
  def __init__(self, id, name ,marks):
      self.id = id
      self.name = name
      self.marks = marks

  # Methods (Functions)
  def display(self):
      print("ID: ", self.id)
      print("Name: ", self.name)
      print("Marks: ", self.marks)

# Object creation
s1 = Student(101, "Alice", 95)
s1.display()
```
