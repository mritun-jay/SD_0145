```python
import time
from IPython.display import clear_output
from datetime import datetime

while True:
    clear_output(wait=True)

    current_time = datetime.now().strftime("%I:%M:%S %p")

    print("   DIGITAL CLOCK")
    print(current_time)

    time.sleep(1)
```

```python
from datetime import datetime
import pytz

# Define IST timezone
ist = pytz.timezone('Asia/Kolkata')

# Get current time in IST
current_time = datetime.now(ist)

# Print in 12-hour format with AM/PM
print("12-hour format:", current_time.strftime("%I:%M %p"))

# Print in 24-hour format
print("24-hour format:", current_time.strftime("%H:%M"))
```

```python
import time
from IPython.display import clear_output
from datetime import datetime
from zoneinfo import ZoneInfo

while True:
  clear_output(wait=True)

  ist_time = datetime.now(ZoneInfo("Asia/Kolkata"))

  print("   Digital Clock")
  print("Date :",ist_time.strftime("%d-%m-%Y"))
  print("Time :",ist_time.strftime("%I:%M:%S %p"))
  print("Zone :","Indian Standard Time (IST")
  print("Day :",ist_time.strftime("%A"))

  time.sleep(1)
```
